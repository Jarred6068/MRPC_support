```python
import pandas as pd
import numpy as np
import re
import scimpute
```

```python
df = scimpute.read_data_into_cell_row('10x_human_pbmc_68k.nz40.msk90.saver.csv.gz', 'gene_row')
```

```python
df.head()
```

```python
change = lambda x: re.sub('\.', '-', x)
```

```python
new_index = list(map(change, df.index))
```

```python
df.index = new_index
```

```python
df = df.transpose()  # into gene_row like other saver.csv to avoid confusion
```

```python
round(df, 4)
```

```python
df.to_csv('saver.fixed.gene_row.csv')
```
